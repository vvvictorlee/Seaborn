```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style='whitegrid')
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
iris.info()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))

sns.swarmplot(data=iris, x='Species', y='SepalLengthCm', ax=ax1)
sns.pointplot(data=iris, x='Species', y='SepalLengthCm', ax=ax1)

sns.swarmplot(data=iris, x='Species', y='SepalLengthCm', ax=ax2, color='black', alpha=0.5)
sns.boxplot(data=iris, x='Species', y='SepalLengthCm', ax=ax2)

sns.countplot(data=iris, x='Species', ax=ax3)
```

```python
sns.pairplot(data=iris.drop('Id', axis=1), hue='Species', 
             diag_kind='kde', diag_kws=dict(shade=True))
```

```python
sns.jointplot(data=iris, x='SepalLengthCm', y='SepalWidthCm')
```

```python
sns.jointplot(data=iris, x='PetalLengthCm', y='PetalWidthCm')
```
